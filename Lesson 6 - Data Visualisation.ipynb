### What are Line Graphs?
Line graphs are used to display data points connected by lines, often to visualise changes over time.

```python
# Creating a sample dataset
years = np.arange(2010, 2021)
sales = np.array([100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300])

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(years, sales, marker='o', linestyle='-', color='b')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Trends in Data
### Identifying Trends: Seasonal, Cyclical, and Random
Trends help in understanding underlying patterns in time series data.

```python
# Creating a time series dataset with seasonal and random trends
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=365)
sales = np.sin(np.linspace(0, 3 * np.pi, 365)) * 100 + np.random.normal(0, 10, 365) + 200

plt.figure(figsize=(12, 6))
plt.plot(dates, sales, label='Sales')
plt.title('Seasonal Trends in Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Scatter Plots
### Introduction to Scatter Plots
Scatter plots are used to visualize the relationship between two numerical variables.

```python
# Generating random data
np.random.seed(0)
x = np.random.rand(100)
y = x * 3 + np.random.normal(0, 0.1, 100)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='g')
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Correlation and Causation
### Understanding Correlation
Correlation measures the strength and direction of the relationship between two variables.

```python
# Enhanced synthetic data to show clearer trends
np.random.seed(42)
data_size = 100

# Time variable affecting both variables similarly
time = np.linspace(1, 12, data_size)

# Ice cream sales - higher variance to make the trend clearer
ice_cream_sales = 50 + 2.5 * time + np.random.normal(0, 20, data_size)

# Sunglasses sales - directly correlated with time and similar trend
sunglasses_sales = 30 + 2.5 * time + np.random.normal(0, 20, data_size)

# Generating trend lines
ice_cream_trend = np.poly1d(np.polyfit(time, ice_cream_sales, 1))(time)
sunglasses_trend = np.poly1d(np.polyfit(time, sunglasses_sales, 1))(time)

plt.figure(figsize=(12, 8))
plt.scatter(time, ice_cream_sales, color='purple', alpha=0.6, label='Ice Cream Sales')
plt.plot(time, ice_cream_trend, "r--", label='Trend: Ice Cream Sales')
plt.scatter(time, sunglasses_sales, color='orange', alpha=0.6, label='Sunglasses Sales')
plt.plot(time, sunglasses_trend, "b--", label='Trend: Sunglasses Sales')
plt.title('Clear Trends Showing Correlation Without Causation')
plt.xlabel('Time (Months)')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

## Finding Outliers
### Identifying Outliers in Data
Outliers are data points that differ significantly from other observations.

```python
# Adding outliers to the dataset
x_with_outliers = np.append(x, [0.1, 0.2])
y_with_outliers = np.append(y, [3, 5])

plt.figure(figsize=(10, 6))
plt.scatter(x_with_outliers, y_with_outliers, color='r')
plt.title('Scatter Plot with Outliers')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Bar Plots
### Introduction to Bar Plots
Bar plots are used to compare the frequency or magnitude of different categories.

```python
# Creating a sample dataset
categories = ['Category A', 'Category B', 'Category C']
values = [25, 35, 30]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='orange')
plt.title('Bar Plot of Categories')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

## Frequency Tables and Histograms
### Frequency Tables
Frequency tables summarize the distribution of a categorical variable.

```python
# Creating a frequency table
data = pd.Series(['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'C', 'C'])
frequency_table = data.value_counts()

print("Frequency Table:\n", frequency_table)
```

### Histograms
Histograms visualize the distribution of a continuous variable by dividing data into bins.

```python
# Generating random data and plotting a histogram
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Types of Distribution
### Normal, Skewed, and Uniform Distributions
Different types of distributions indicate how data points are spread.

```python
# Normal Distribution
normal_data = np.random.normal(0, 1, 1000)

# Skewed Distribution
skewed_data = np.random.exponential(1, 1000)

# Uniform Distribution
uniform_data = np.random.uniform(-3, 3, 1000)

plt.figure(figsize=(14, 8))

plt.subplot(3, 1, 1)
plt.hist(normal_data, bins=30, color='blue', alpha=0.7)
plt.title('Normal Distribution')

plt.subplot(3, 1, 2)
plt.hist(skewed_data, bins=30, color='green', alpha=0.7)
plt.title('Skewed Distribution')

plt.subplot(3, 1, 3)
plt.hist(uniform_data, bins=30, color='red', alpha=0.7)
plt.title('Uniform Distribution')

plt.tight_layout()
plt.show()
```
